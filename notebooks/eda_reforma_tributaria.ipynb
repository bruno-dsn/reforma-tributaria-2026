```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Criando os dados da Reforma Tributária 2026
dados = {
    'setor': [
        'Cesta Básica',
        'Alimentos (geral)',
        'Medicamentos',
        'Serviços de Saúde',
        'Educação',
        'Produtos de Higiene',
        'Serviços Profissionais',
        'Transporte',
        'Comércio Geral',
        'Construção Civil',
        'Tecnologia',
        'Bares e Restaurantes',
        'Hotéis e Turismo',
        'Bebidas Alcoólicas',
        'Bebidas Açucaradas'
    ],
    'aliquota_atual': [
        8.0, 15.8, 19.5, 22.0, 17.5, 23.0, 18.0, 25.0, 27.5, 30.0, 20.0, 28.0, 26.0, 40.0, 35.0
    ],
    'aliquota_nova': [
        0.0, 10.6, 10.6, 10.6, 10.6, 10.6, 18.5, 26.5, 26.5, 26.5, 26.5, 15.9, 15.9, 31.5, 28.5
    ]
}

df = pd.DataFrame(dados)
```

```python
# Salvando o CSV no ambiente do Colab
df.to_csv('aliquotas_reforma_tributaria.csv', index=False)

df.head()
```

```python
# Criando a variação de alíquota
df['variacao'] = df['aliquota_nova'] - df['aliquota_atual']

# Estatísticas descritivas
df.describe()
# Ordenando por maior impacto
df.sort_values('variacao')
```

```python
plt.figure(figsize=(8, 6))
plt.barh(df['setor'], df['variacao'])
plt.title('Variação de Alíquotas por Setor – Reforma Tributária 2026')
plt.xlabel('Variação (%)')
plt.ylabel('Setor')
plt.show()
```
